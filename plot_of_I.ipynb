```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import gamma
from scipy import integrate
from scipy.constants import c, m_e, epsilon_0 as epsi_0, elementary_charge as q

from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
%matplotlib inline

def sech(x):
    return 1 / np.cosh(x)

def gauss_to_tesla(gauss):
    return gauss * 1e-4

def degrees_to_radians(degrees):
    return degrees * np.pi / 180

def plot_data(sphase, I, Khi, Delta, title):
    plt.figure(figsize=(12, 5))
    plt.plot(sphase, I)
    plt.title(title)
    plt.xlabel('sphase')
    plt.ylabel('I')
    plt.show()
    plt.close()

# Constants
f = 0.008538220
BL_G = 0.3
p = 2
N0 = 10
K0 = 10
G_v = 10
tobs = 0

phi = np.pi / 3
theta = np.pi / 2
zeta = np.pi / 6 
Omega = 2 * np.pi * f 
rL = 2.99792458e+08 / Omega
Bv = np.sqrt(G_v**2 - 1) / G_v

BL_T = gauss_to_tesla(BL_G)

r = np.linspace(rL, (10**8)*rL, 1000)
z = r / rL

increment = 10**(-3)
max_tret = 2 * np.pi / Omega

tret = np.arange(0, max_tret, increment)
sphase = Omega * tret / (2 * np.pi)

D_v = (1 / G_v) * (1 - Bv * (np.cos(theta) * np.cos(zeta) +
                             np.sin(theta) * np.sin(zeta) * np.cos(phi)))

u1 = (p - 1) / 2 
u2 = (p + 3) / 2 
u3 = (p + 1) / 4 
u4 = (3*p + 7)/12 
u5 = (3*p - 1)/12 
u6 = (p + 7/3)/(p + 1)

GE_u4 = gamma(u4)
GE_u5 = gamma(u5)
Kp = (np.sqrt(3) * GE_u4 * GE_u5 * abs(q)**3) / \
     (32 * np.pi**2 * epsi_0 * m_e * c) * \
     ((3 * abs(q)) / (m_e**3 * c**4))**u1

Khi_values = np.arange(60, 81, 10)
Delta_values = np.linspace(7, 9, 3)

for Khi in Khi_values:
    Khi_r = degrees_to_radians(Khi)
    psi = np.cos(theta) * np.cos(Khi_r) + np.sin(theta) * np.sin(Khi_r) * np.cos(phi + z*(1 - Bv*np.sin(zeta)) / Bv)
    
    for Delta in Delta_values:
        B = Bv * BL_T * 1/z * np.tanh(psi/Delta) * np.sqrt(np.sin(Khi)**2 + G_v**2 * Bv**2 * 1/z**2)
        
        x = np.multiply(psi, Delta)
        sech_x = sech(x)
        sech2 = np.power(sech_x, 2)
        r2Ke = N0 * ((K0 - 1) * sech2 + 1)

        r2s0 = Kp * r2Ke * (Omega**-u1) * (D_v**u2) * (((B/G_v)**2) * (1 - (D_v *(-np.sin(phi)*np.sin(zeta)))**2))**u3
        
        # Integrate
        def integrand(r, theta, phi):
            return np.sin(theta)
        
        value, err = integrate.tplquad(integrand, 0, np.inf, 0, np.pi, 0, 2 * np.pi)
        
        # Ensure value is a scalar
        if np.isscalar(value):
            I = np.multiply(r2s0, value)
        else:
            I = np.multiply(r2s0, np.sum(value))

        # Ensure length of I matches length of sphase
        if len(I) != len(sphase):
            I = np.interp(sphase, np.linspace(0, 1, len(I)), I)
        
        title = f"Plot of I for $\Gamma_v$={G_v}, $K_0$={K0} ,\n p={p}, Khi={Khi} and Delta = {Delta}"
        plot_data(sphase, I, Khi, Delta, title)
```
