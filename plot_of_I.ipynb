# Plot of I using Scipy Integral module

\begin{align*}
I(t_{obs}) &= \int_{0}^{+\infty}\int_{0}^{\pi}\int_{0}^{2\pi}\; s_o(r, t_{ret})\left(\frac{p + \frac{7}{3}}{p + 1 } \right) \: r^2 \sin(\theta)\;dr\;d\theta\;d\varphi \\ 
&=\left(\frac{p + \frac{7}{3}}{p + 1 } \right)\int_{0}^{+\infty}\int_{0}^{\pi}\int_{0}^{2\pi}\; r^2 \: s_o(r, t_{ret}) \:  \sin(\theta)\;dr\;d\theta\;d\varphi \\
\end{align*}<br>

$$ r^2s_o(r, t_{ret}) =  r^2 \: \kappa \: K(r, t_{ret}) \omega^{-\frac{P - 1}{2}} D_v^\frac{p + 3}{2} \left( \frac{\textbf{B}}{\Gamma_v } \sqrt{ 1 - \left( D_v \textbf{n} \cdot \textbf{b} \right)^2} \right)^\frac{P + 1}{2} $$ 

$$ r^2s_o(r, t_{ret}) = \kappa \:  r^2K(r, t_{ret}) \omega^{-\frac{P - 1}{2}} D_v^\frac{p + 3}{2} \left[ \frac{\textbf{B}^2}{\Gamma_v^2 } \left( 1 - \left( D_v \textbf{n} \cdot \textbf{b} \right)^2 \right) \right]^{\frac{P + 1}{2} \cdot \frac{1}{2}} $$ <br>

- $ u_1 = \frac{P - 1}{2} $ <br><br>
- $ u_2 = \frac{P + 3}{2} $ <br><br>
- $ u_3 = \frac{P + 1}{4} $ <br><br>

$$ r^2s_o(r, t_{ret}) = \kappa \:  r^2 \: K(r, t_{ret}) \omega^{- u_1} D_v^{u_2} \left[ \frac{\textbf{B}^2}{\Gamma_v^2 } \left( 1 - \left( D_v \textbf{n} \cdot \textbf{b} \right)^2 \right) \right]^{u_3} $$ <br>

- $ \kappa = \frac{\sqrt{3}}{2\pi} \:  \frac{1}{4} \: \Gamma_{eu}\left( \frac{3p + 7}{12} \right) \Gamma_{eu}\left( \frac{3p - 1}{12} \right) \frac{\mid q \mid^3}{4 \pi \epsilon_0 m c} \left(\frac{3\mid q \mid}{m^3 c^4} \right)^{\frac{P - 1}{2}}  $ <br>

    - $ u_4 = \frac{3p + 7}{12} $ <br><br>
    - $ u_5 = \frac{3p - 1}{12} $ <br><br>

$$ \kappa =\frac{\sqrt{3} \mid q \mid^3}{32 \pi^2 \epsilon_0 m c}  \Gamma_{eu}(u_4) \Gamma_{eu}(u_5) \left(\frac{3\mid q \mid}{m^3 c^4} \right)^{u_1} $$ <br>

- $ K(r, t_{ret}) = \frac{(N - N_0) \: sech^2(\Delta \cdot \psi) + N_0}{r^2}$ <br><br>
  - $K_0 = \frac{N}{N_0} $ $ \Rightarrow  N = K_0 \: N_0 $ <br><br>
$$r^2K(r, t_{ret}) = N_0 \left[(K_0 - 1)sech^2(\Delta \cdot \psi) + 1 \right] $$<br>

The expression of $ \psi $ is given by the **Eq.(2) in Petri 2013**:

$\rhd$ $\psi = \cos(\theta) \: \cos(\chi) + \sin(\theta) \: \sin(\chi) \cos(\varphi - \Omega \: (t_{ret} -\frac{r}{\beta_v \: c}) )$


Where $\quad \beta_v = \frac{v}{c} \quad$ and $\quad  \Omega = \frac{c}{r_L} \quad$  <br><br>

\begin{align*}
\psi &=  \cos(\theta) \: \cos(\chi) + \sin(\theta) \: \sin(\chi) \cos(\varphi - \frac{c}{r_L} \: (t_{ret}  -\frac{r}{\beta_v \: c} ) \\
&= \cos(\theta) \: \cos(\chi) + \sin(\theta) \: \sin(\chi) \cos \left[\varphi + \frac{c}{r_L} \: \frac{r}{c \beta_v} ( - \frac{\beta_v \; c}{r} \; t_{ret} + 1 )\right] \\
&= \cos(\theta) \: \cos(\chi) + \sin(\theta) \: \sin(\chi) \cos \left[\varphi + \frac{r}{\beta_v r_L} \: ( 1 - \frac{\beta_v \; c}{r}  \; t_{ret})\right] \\
\end{align*}

$\rhd$ $ \psi = \cos(\theta) \: \cos(\chi) + \sin(\theta) \: \sin(\chi) \cos \left[\varphi + \frac{r}{\beta_v \; r_L} \: ( 1 - \frac{\beta_v \; c}{r}  \; t_{ret})\right]$ <br><br>

- $D_v = \frac{1}{\Gamma_v \left( 1 - \boldsymbol{\beta} \cdot \textbf{n} \right)}$ <br><br>
$\boldsymbol{\beta} \cdot \textbf{n} = \: \beta \: \mu \quad with \quad \mu = cos(\theta) \: cos(\zeta) + sin(\theta) \: sin(\zeta) \: cos(\varphi) $ <br><br>
$\boldsymbol{\beta} \cdot \textbf{n}= \beta \left[ \: cos(\theta) \: cos(\zeta) + sin(\theta) \: sin(\zeta) \: cos(\varphi) \right] $<br>
     
$$ D_v = \frac{1}{\Gamma_v (1 - \beta \left[ \: cos(\theta) \: cos(\zeta) + sin(\theta) \: sin(\zeta) \: cos(\varphi) \right])}$$
<br><br><br>
- According the equation $Eq(5) \: and \: Eq(9) \: and \: your \: proposition \: we \: will \: obtain \:B$ <br>

\begin{align*}
\textbf{B'} &= \frac{\textbf{B}^2}{\Gamma_v^2} \\
\textbf{B'} &= \beta_v^2 \; B_L^2 \; \frac{r_L^2}{r^2} \left( \frac{sin^2(\upsilon)}{\Gamma_v^2} + \beta_v^2 \frac{r_L^2}{r^2} \right) tanh^2 \left( \psi / \Delta \right)\\
\textbf{B}^2 &= \beta_v^2 \; B_L^2 \; \frac{r_L^2}{r^2} \; \left( sin^2(\upsilon) +  \frac{\Gamma_v^2 \; \beta_v^2 \; r_L^2}{r^2}  \right) tanh^2 \left( \psi / \Delta \right)\\
\textbf{B} &= \frac{r_L}{r} \; \beta_v \; B_L \; tanh \left( \psi / \Delta \right) \sqrt{ \left[ sin^2(\upsilon) + \left( \frac{\Gamma_v \; \beta_v \; r_L}{r} \right)^2 \right]}
\end{align*}<br>


- $\textbf{b} \cdot \textbf{n} \quad \text{with} \quad \textbf{b} = \{- sin(\varphi),\: cos(\varphi),\: 0\} \quad and \quad \textbf{n} = \{ sin(\zeta),\: 0,\: cos(\zeta) \}$ <br>

$ \qquad \textbf{b} \cdot \textbf{n} = - sin(\varphi) \: sin(\zeta)$ <br><br>

\begin{align*}
I(t_{obs}) &=\left(\frac{p + \frac{7}{3}}{p + 1 } \right)\int_{0}^{+\infty}\int_{0}^{\pi}\int_{0}^{2\pi}\; \kappa \:  r^2 \: K(r, t_{ret}) \omega^{- u_1} D_v^{u_2} \left[ \frac{\textbf{B}^2}{\Gamma_v^2 } \left( 1 - \left( D_v \textbf{n} \cdot \textbf{b} \right)^2 \right) \right]^{u_3} \:  \sin(\theta)\;dr\;d\theta\;d\varphi \\
\end{align*}

- $ u_6 = \left(\frac{p + \frac{7}{3}}{p + 1 } \right)$ <br><br>

$$ I(t_{obs}) = u_6 \: \int_{0}^{+\infty}\int_{0}^{\pi}\int_{0}^{2\pi}  \: \kappa \;  r^2 \: K(r, t_{ret}) \; \omega^{- u_1} \; D_v^{u_2} \left[ \frac{\textbf{B}^2}{\Gamma_v^2 } \left( 1 - \left( D_v \textbf{n} \cdot \textbf{b} \right)^2 \right) \right]^{u_3} \:  \sin(\theta)\;dr\;d\theta\;d\varphi$$

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import integrate
from scipy.special import gamma
from scipy.constants import c, m_e, epsilon_0 as epsi_0, elementary_charge as q

def sech(x):
    return 1 / np.cosh(x)

def gauss_to_tesla(gauss):
    return gauss * 1e-4

def degrees_to_radians(degrees):
    return degrees * np.pi / 180

def plot_data(sphase, I_values, Khi, Delta, title):
    plt.figure(figsize=(12, 5))
    plt.plot(sphase, I_values)
    plt.title(title)
    plt.xlabel('sphase')
    plt.ylabel('I')
    plt.show()
    plt.close()

# Constants
f = 0.008538220
BL_G = 0.3
p = 2
N0 = 10
K0 = 10
G_v = 10

zeta = np.pi / 6 
Omega = 2 * np.pi * f 
rL = 2.99792458e+08 / Omega
Bv = np.sqrt(G_v**2 - 1) / G_v

BL_T = gauss_to_tesla(BL_G)

increment =  0.117
max_tret = 2 * np.pi / Omega

tret = np.arange(0, max_tret, increment)
sphase = Omega * tret / (2 * np.pi)

u1 = (p - 1) / 2 
u2 = (p + 3) / 2 
u3 = (p + 1) / 4 
u4 = (3*p + 7)/12 
u5 = (3*p - 1)/12 
u6 = (p + 7/3)/(p + 1)

GE_u4 = gamma(u4)
GE_u5 = gamma(u5)
Kp = (np.sqrt(3) * GE_u4 * GE_u5 * abs(q)**3) / \
     (32 * np.pi**2 * epsi_0 * m_e * c) * \
     ((3 * abs(q)) / (m_e**3 * c**4))**u1

Khi_values = np.arange(60, 81, 10)
Delta_values = np.linspace(7, 9, 3)

for Khi in Khi_values:
    Khi_r = degrees_to_radians(Khi)
    
    for Delta in Delta_values:
        # Pre-compute values outside the integrand
        cos_Khi_r = np.cos(Khi_r)
        sin_Khi_r = np.sin(Khi_r)
        
        I_values = []
        
        for tr in tret:
            def integrand(r, theta, phi):
                D_v = 1 / ( G_v * (1 - Bv * (np.cos(theta) * np.cos(zeta) + 
                                 np.sin(theta) * np.sin(zeta) * np.cos(phi) ) ) )
                
                
                psi = cos_Khi_r * np.cos(theta) + sin_Khi_r * np.sin(theta) * np.cos(phi + (r / rL * Bv) * (1 - (Bv * c * tr) / r))
                
                B = rL / r * Bv * BL_T * np.tanh(psi / Delta) * np.sqrt(sin_Khi_r**2 + ((G_v * Bv * rL) / r)**2)
                
                sech2 = sech(psi * Delta)**2
                r2Ke = N0 * ((K0 - 1) * sech2 + 1)
                
                r2s0 = Kp * r2Ke * (Omega**-u1) * (D_v**u2) * ( ( (B / G_v)**2 ) * ( 1 - ( D_v * ( - np.sin(phi) * np.sin(zeta) ) )**2 ) )**u3
                
                return  r2s0 * np.sin(theta)
                
            I, _ = integrate.tplquad(integrand, 0, np.inf, 0, np.pi, 0, 2 * np.pi)
            I *= u6
            I_values.append(I)
        
        title = f"Plot of I for $\Gamma_v$={G_v}, $K_0$={K0} ,\n p={p}, Khi={Khi} and Delta = {Delta}"
        plot_data(sphase, I_values, Khi, Delta, title)
```
