```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy import integrate
from scipy.constants import c, m_e, epsilon_0 as epsi_0, elementary_charge as q

from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
%matplotlib inline

def sech(x):
    return 1 / np.cosh(x)

def gauss_to_tesla(gauss):
    """
    Convert Gauss to Tesla.

    Parameters:
        gauss : Value in Gauss.

    Returns:
        Value converted to Tesla.
    """
    return gauss * 1e-4

def degrees_to_radians(degrees):
    """
    Convert degrees to radians.

    Parameters:
        degrees: Value in degrees.

    Returns:
        Value converted to radians.
    """
    return degrees * np.pi / 180

def plot_data(z, I, Khi, Delta, title):
    """
    Plot I for a given Khi and Delta value.

    Parameters:
        z: Array of z values.
        I: Array of the Intensity
        Khi: Value of Khi.
        Delta: Value of Delta.
    """
    plt.figure(figsize=(12, 5))
    plt.plot(z, I)
    plt.title(title)
    plt.xlabel('r/rL')
    plt.ylabel('I')
    plt.show()
    plt.close()

    
# Constants
f = 0.008538220
BL_G = 0.3

p = 2
N0 = 10
K0 = 10
G_v = 10

phi = np.pi / 3
theta = np.pi / 2
Omega = 2 * np.pi * f 
rL = 2.99792458e+08 / Omega
Bv = np.sqrt(G_v**2 - 1) / G_v

D_v = 1/G_v * (1 - Bv)
BL_T = gauss_to_tesla(BL_G)

r = np.linspace(rL, (10**8)*rL, 1000)
z = r / rL


u1 = (p - 1) / 2 #u3
u2 = (p + 3) / 2 #u4
u3 = (p + 1) / 4 #u5
u4 = (3*p + 7)/12 #u1
u5 = (3*p - 1)/12 #u2
u6 = (p + 7/3)/(p + 1)


GE_u4 = gamma(u4)
GE_u5 = gamma(u5)
Kp = (np.sqrt(3) * GE_u4 * GE_u5 * abs(q)**3)/(32 *  np.pi**2  * epsi_0 * m_e * c)*((3 * abs(q))/(m_e**3 * c**4))**u1


# Varying parameters
Khi_values = np.arange(60, 81, 10)  # Vary Khi from 60 to 80 with step 10 degrees
Delta_values = np.linspace(7, 9, 3)  # Vary Delta from 7 to 9

for Khi in Khi_values:
    Khi_r = degrees_to_radians(Khi)
    psi = np.cos(theta) * np.cos(Khi_r) + np.sin(theta) * np.sin(Khi_r) * np.cos(phi + z*(1 - Bv) / Bv)
    
    for Delta in Delta_values:
        br = Bv**2 * BL_T * 1 / z**2 * np.tanh(psi / Delta)
        bphi = - Bv * BL_T * (1 / z) * np.sin(theta)  * np.tanh(psi / Delta)
        B = Bv * BL_T * 1/z * np.tanh(psi/Delta) * np.sqrt(np.sin(Khi)**2 + G_v**2 * Bv**2 * 1/z**2)
        
        x = np.multiply(psi, Delta)
        sech_x = sech(x)
        sech2 = np.power(sech_x, 2)
        r2Ke = N0 * ((K0 - 1) * sech2 + 1)

        r2s0 = Kp * r2Ke * (Omega**-u1) * (D_v**u2) * (((B/G_v)**2) * (1 - (D_v *(br + bphi))**2))**u3
        f4 = lambda r1, theta1, phi1: np.sin(theta1) 
        value, err = integrate.tplquad(f4,0,np.inf,0,np.pi, 0, 2*np.pi)    

        I = np.multiply(r2s0, value)
        title = f"Plot of I for $\Gamma_v$={G_v}, $K_0$={K0} ,\n p={2}, Khi={Khi} and Delta = {Delta}"
        plot_data(z, I, Khi, Delta, title)
```
